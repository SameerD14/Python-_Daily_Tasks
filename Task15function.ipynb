##### Here are 5 Hard Questions on Python Functions
##### 1. (Hard) Write a function process_data(*args, **kwargs) that:
##### • Takes any number of positional arguments (numbers)
##### • Takes keyword arguments where keys are operation names ("add", "multiply","power")
##### • Performs the requested operations on all numbers
##### • Returns a dictionary with results Example: process_data(2, 3, 4, add=True,multiply=True) should return appropriate results.

```python
def process_data(*args, **kwargs):
    result={}
    if kwargs.get("add"):
        total=0
        for i in args:
            total+=i
        result["add"]=total
    if kwargs.get("multiply"):
        product=1
        for i in args:
            product*=i
        result["multiply"]=product
    return result
    
process_data(2,3,4,add=True,multiply=True)
```

##### 2. (Hard) Create a function make_multiplier(n) that returns another function which multiplies any number by n. (Use closure concept). Then create double = make_multiplier(2) and triple = make_multiplier(3) and test them.

```python
def make_multiplier(n):
    def make_multiplier1(x):
        return n*x  
    return make_multiplier1
double=make_multiplier(2)
triple=make_multiplier(3)
print("Double of 2=",double(2))
print("Triple of 8=",triple(8))
    
        
```

##### 3. (Hard) Write a function smart_divide(func) as a decorator that:
##### • Takes a function as argument
##### • Checks if the second argument is zero before division
##### • If zero, prints "Cannot divide by zero!" and returns None
##### • Otherwise, calls the original function

```python
def smart_divide(divisor):
    def divide(dividend):
        if divisor == 0:
            return "Cannot divide by zero!"
        return dividend / divisor
    return divide
divide_by_2 = smart_divide(2)
divide_by_0 = smart_divide(0)
print(divide_by_2(10))
print(divide_by_0(10))
```
