##### 15. Take three sides of a triangle. Check if they form a valid triangle. If,yes, also print if it is Equilateral, Isosceles, or Scalene.

```python
a=int(input("Enter a side a =" ))
b=int(input("Enter a side b =" ))
c=int(input("Enter a side c =" ))
if (a+b)>=c and (b+c)>=a and (a+c)>=b:
    print("Valid Traingle")
    if (a==b and a==c) or (b==c and a==c) or (a==b and b==c):
        print("Equilateral Traingle..")
    elif a==b or b==c or a==c:
        print("Isoceles Traingle")
    else:
        print("Scalene Traingle")
else:
        print("Invalid Traingle")
```

##### 16. Take student marks in 5 subjects. Calculate percentage and give grade: 90+ → A+, 80-89 → A, 70-79 → B, 60-69 → C, below 60 → Fail.

```python
a=int(input("Enter a marks sub1: "))
b=int(input("Enter a marks sub2: "))
c=int(input("Enter a marks sub3: "))
d=int(input("Enter a marks sub4: "))
e=int(input("Enter a marks sub5: "))
total=a+b+c+d+e
per=total*100/500
if per>=90:
    print("A+")
elif per>=80:
    print("A")
elif per>=70:
    print("B")
elif per>=60:
    print("C")
else:
    print("Fail")
```

##### 20. Take three numbers. Print them in ascending order using if-elif-else(without using sort()).
