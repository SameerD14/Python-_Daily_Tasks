##### 19. Given data = {"a": 10, "b": 20, "c": 10}, use items() and print only keys where value is 10.

##### 25. (Hard) Given data = (10, 20, 30, 20, 40, 20, 50), convert it into a dictionary where key is the number and value is its count of occurrences (using tuple methods + dictionary)

```python
data = (10, 20, 30, 20, 40, 20, 50)
dict1={}
for i in data:
    dict1.update({i:data.count(i)})
print(dict1)
    
```
