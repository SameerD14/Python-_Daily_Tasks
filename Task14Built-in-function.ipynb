##### 3. Advanced Enumeration + Zip + Condition You have two lists:Python
##### names = ["Sumit", "Rahul", "Priya", "Aman", "Neha"]
##### scores = [85, 92, 78, 95, 88]
##### Write a single list comprehension using enumerate() and zip() that returns a list oftuples in this format: [(index, name, "Excellent")] if score ≥ 90, else [(index, name,"Good")]

```python
names = ["Sumit", "Rahul", "Priya", "Aman", "Neha"]
scores = [85, 92, 78, 95, 88]
result = [(index, name, "Excellent" if score >= 90 else "Good")
          for index, (name, score) in enumerate(zip(names, scores))]
print(result)
```
