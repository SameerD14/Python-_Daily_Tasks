##### 30. Interview Challenge
##### students = [{"name":"Ram","marks":[70,80,90]},{"name":"Amit","marks":[90,92,94]},{"name":"Neha","marks":[60,65,70]},{"name":"Raj","marks":[88,84,90]}]
##### Using only lambda with built-in functions (map, filter, sorted, max, min, reduce where appropriate):

##### 1. Calculate each student's average marks.

```python
from functools import reduce
students = [{"name":"Ram","marks":[70,80,90]},{"name":"Amit","marks":[90,92,94]},{"name":"Neha","marks":[60,65,70]},{"name":"Raj","marks":[88,84,90]}]
result=list(map(lambda student:{"name":student["name"],"average":reduce(lambda x,y:x+y,
                                                                        student["marks"])/len(student["marks"])},students))
print(result)
```

##### 2. Sort students by average marks (highest first).

```python
sort=sorted(result,reverse=True,key=lambda x:x["average"])
print(sort)
```

##### 3. Find the topper.

```python
topper=max(result,key=lambda x:x["average"])
print(topper)
```

##### 4. Filter students whose average is greater than 80.

```python
filtered=list(filter(lambda x:x["average"]>80,result))
print(filtered)
```

##### 5. Print only the names of the filtered students.

```python
name=list(map(lambda x:x["name"],filtered))
print(name)
```
