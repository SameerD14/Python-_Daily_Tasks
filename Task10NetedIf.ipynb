##### 18.Take three sides of a triangle. Check if valid, then check if it is a right-angled triangle using Pythagoras theorem.

```python
a = int(input("Enter first side: "))
b = int(input("Enter second side: "))
c = int(input("Enter third side: "))
if (a+b>c) and (a+c>b) and (b+c >a):
    largest = max(a, b, c)
    if largest == a:
        if a*a == b*b + c*c:
            print("Right Angle Triangle")
        else:
            print("Not a Right Angle Triangle")
    elif largest == b:
        if b*b == a*a + c*c:
            print("Right Angle Triangle")
        else:
            print("Not a Right Angle Triangle")
    else:
        if c*c == a*a + b*b:
            print("Right Angle Triangle")
        else:
            print("Not a Right Angle Triangle")
else:
    print("Invalid Triangle")
```
