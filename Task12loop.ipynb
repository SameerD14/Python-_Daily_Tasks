##### 4.Write a program to reverse a number using a while loop.

```python
num=int(input("Enter a number:"))
reverse=0
print("Original no=",num)
while num>0:
    digit=num%10
    reverse=reverse*10+digit
    num=num//10
print("Reverse no=",reverse)
```

##### 6.Write a program to check whether a number is palindrome or not.

```python
num=int(input("Enter a number:"))
temp=num
reverse=0
while num>0:
    digit=num%10
    reverse=reverse*10+digit
    num//=10
if reverse==temp:
    print(temp,"Palindrome number")
else:
    print(temp,"Not a palindrome number")
```

##### 7.Write a program to check whether a number is Armstrong number.

```python
n=int(input("Enter a number:"))
num=n
count=0
while num>0:
    count+=1
    num//=10
num=n
total=0
while num>0:
    digit=num%10
    total=total+digit**count
    num//=10
if n==total:
    print(n,"is a armstrong number")
else:
    print(n,"is not a armstrong number")
    
```

##### 12.Write a program to find factorial of a number.

```python
num=int(input("Enter a number:"))
factorial=1
i=num
while i>0:
    factorial=factorial*i
    i-=1
print(f"Factorial of {num}={factorial}")
```

```python
balance=10000
print(" To check balance press 1 \n To deposit Money press 2 \n To withdraw money press 3 \n To Exit press 4")
while True:
    choice=int(input("Enter a number:"))
    if choice==1:
        print("balance=",balance)
        
    elif choice==2:
        deposit=int(input("Enter a deposit amount="))
        balance+=deposit
        print("Current balance is ",balance)
    elif choice==3:
        withdraw=int(input("Enter a withdraw amount="))
        if withdraw<=balance:
            balance-=withdraw
            print("Current balance is ",balance)
        else:
            print("Insufficient Balance")
    elif choice==4:
         break
    else:
        print("Invalid choice")
```

##### 16.Check leap year repeatedly until user types "stop".

```python
leap_year=[]
while True:
    user=int(input("Enter a number: "))
    if ((user%4==0 and user%100!=0) or user%400):
        leap_year.append(user)
    elif user==0:
        break
print("Leap year=",leap_year)
```

##### 19. Keep taking student marks until negative number entered.

```python
total = 0
count = 0
while True:
    marks = int(input("Enter marks: "))
    if marks < 0:
        break
    total += marks
    count += 1
average = total / count
print("Total Marks =", total)
print("Total Students =", count)
print("Average =", average)
```
