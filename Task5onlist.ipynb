##### 18. Take 5 city names as input from user and store them in a list. Then print them in reverse order.

```python
cities=[]
for i in range(5):
    city=input(f"Enter a city {i+1}=")
    cities.append(city)
cities.reverse()
print("Cities in reverse Order..")
for city in cities:
    print(city)
```

##### 19. Given data = [10, 20, 30, 20, 40, 20], remove all occurrences of 20 (use loop or multiple remove).

```python
number=[10, 20, 30, 20, 40, 20]
count=number.count(20)
for i in range(count):
    number.remove(20)
print("Using loop: ",number)
# Using Multile Remove is
number=[10, 20, 30, 20, 40, 20]
number.remove(20)
number.remove(20)
number.remove(20)
print("Using Multiple reverse=",number)
```
