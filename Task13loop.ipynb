##### 11. Reverse a number using while loop.

```python
num=int(input("Enter a number:"))
reverse=0
while num>0:
    digit=num%10
    reverse=reverse*10+digit
    num=num//10
print("reverse number=",reverse)
```

##### 22. Print a hollow square pattern of size n using nested loop.

```python
n=int(input("Enter a number:"))
for i in range(1,n+1):
    for j in range(1,n+1):
        if i==1 or i==n or j==1 or j==n:
            print("*",end=" ")
        else:
            print(" ",end=" ")
    print()
```

##### 23. Print Floyd’s Triangle using nested for loop.

```python
n=int(input("Enter a rows="))
k=1
for i in range(1,n+1):
    for j in range(1,i+1):
        print(k,end=" ")
        k=k+1
    print()
```
