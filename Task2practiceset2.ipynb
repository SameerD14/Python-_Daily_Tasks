```python
# Take age and marks from user.
# Check if the student is eligible for admission:
# Age must be between 18 and 25 AND marks >= 60

age=int(input("Enter the age of student: "))
marks=int(input("Enter the marks of student: "))

if age>=18 and age<=25 and marks>=60:
    print("Eligible for addmision")
else:
    print("Not Elligible for addmission") 
    
```

```python
# Take a number from user.
# Check if the number is between 50 and 100 (using 'and')
# Also check if it is NOT divisible by 5 (using 'not')


num=int(input("enter the number: "))

if num>=50 and num<=100 and not (num%5==0):
    print("the number is between 50 and 100 and not divisible by 5")
else:
    print("the condition not satisfied")
```

```python
# Take three numbers from user.
# Print the largest number using 'and' and 'or' operators.

num1=int(input("enter the first number: "))
num2=int(input("enter the second number:"))
num3=int(input("enter the third number: "))

if num1>=num2 and num1>=num3:
    print("largest num is: ",num1)
elif num2>=num1 and num2>=num3 or num2==num3:
    print("largest num is: ",num2)
else:
    print("largest num is: ",num3)
    
```

```python
# Take a word from user.
# Check if the word is a palindrome or not.

word=input("enter the word:")

if word==word[::-1]:
    print("palindrome")
else:
    print("not palindrome")
    
```

```python
# Take email as input (e.g: student@gmail.com)
# Print only the username (part before @)
# Print only the domain (part after @)


email = input("Enter your email: ")


pos = email.find("@")

print("Username:", email[:pos])
print("Domain:", email[pos + 1:])
```

```python
# Take a password from user.
# Check if password is strong:
# - Length >= 8
# - Contains '@' or '#'
# Print "Strong Password" or "Weak Password"

password=input(" Enter the password:")

if len(password)>= 8 and ("@" in password or "#" in password):
    print("strong password")
else:
    print("weak password")
```

```python
# Take a string from user.
# Check if:
# - The string length is greater than 5
# - AND it starts with 'A' or 'a'
# Print "Valid String" or "Invalid String"

text=str(input(" Enter a string: "))

if len(text)>=5 and (text.startswith("A") or text.startswith("a")):
    print(" Valid string")
else:
    print("Invalid string ")

```
