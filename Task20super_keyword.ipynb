##### 1. Class Name: Shape
##### Problem
##### Create an abstract class named Shape.
##### Explanation
##### A shape can have different formulas for area and perimeter. Therefore, create two abstract
##### methods:
##### • area()
##### • perimeter()
##### Create subclasses like Circle and Rectangle and implement both methods.

```python
from abc import ABC , abstractmethod
class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
    @abstractmethod
    def calculate_perimeter(self):
        pass
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
    def calculate_area(self):
        print("Area=",3.14* (self.radius**2))
    def calculate_perimeter(self):
        print("Perimeter=",2*3.14*self.radius)
class Rectraingle(Shape):
    def __init__(self,length,width):
        self.length=length
        self.width=width
    def calculate_area(self):
        print("Area=",self.length*self.width)
    def calculate_perimeter(self):
        print("Perimeter=",2*(self.length+self.width))  
print("Circle")
cr1=Circle(4)
cr1.calculate_area()
cr1.calculate_perimeter()
print("*"*90)
print("Rectraingle")
rec1=Rectraingle(24,14)
rec1.calculate_area()
rec1.calculate_perimeter()
```

##### Explanation
##### Store the calculator name.
##### __init__(self, name)
##### Create abstract method:
##### • compute(a, b)
##### Create subclass:
##### • ScientificCalculator
##### Use super() to initialize the calculator name.

```python
class Calculator(ABC):
    def __init__(self,name):
        self.name=name
    @abstractmethod
    def compute(self,a,b):
        pass
class ScientificCalculator(Calculator):
    def __init__(self,name):
        super().__init__(name)
    def compute(self,a,b):
        self.a=a
        self.b=b
        if self.name=="ADD":
            result=self.a+self.b
        elif self.name=="SUB":
            result=self.a-self.b
        elif self.name=="MUL":
            result=self.a*self.b
        elif self.name=="DIV":
            result=self.a/self.b
        return result
op=ScientificCalculator("add".upper())
print("opertion=",op.name)
result1=op.compute(8,56)
print(f"a={op.a} ,b={op.b} ={result1}")
op=ScientificCalculator("mul".upper())
print("opertion=",op.name)
result1=op.compute(8,56)
print(f"a={op.a} ,b={op.b} ={result1}")
```

```python
from abc import ABC, abstractmethod
class Report(ABC):
    def __init__(self,title):
        self.title=title
    @abstractmethod
    def generate_report(self):
        pass
    def show_header(self):
        print("Report Title=",self.title)
class PDFReport(Report):
     def __init__(self,title):
        super().__init__(title)
     def generate_report(self):
        self.show_header()
        print("Pdf created Successfully")
class HTMLReport(Report):
     def __init__(self,title):
        super().__init__(title)
     def generate_report(self):
        self.show_header()
        print("HTML report created Successfully")
pdf=PDFReport("Docs")
pdf.generate_report()
print("*"*90)
html=HTMLReport("Header")
html.generate_report()
```
