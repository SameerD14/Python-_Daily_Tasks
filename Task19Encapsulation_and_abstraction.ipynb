##### 1. Class Name: BankAccount Problem
##### Create a class named BankAccount. Explanation
##### A bank account stores money. The balance should not be directly accessible from outside the class, so create a private attribute named __balance. Implement the following methods:
##### • deposit(amount)
##### o Deposit money into the account.
##### o Amount must be greater than 0.
##### • withdraw(amount)
##### o Withdraw money.
##### o Withdrawal should happen only if:
##### ▪ amount > 0
##### ▪ amount <= available balance

```python
class Bank_Account:
    def __init__(self,acc_no,balance):
        self.acc_no=acc_no
        self.__balance=(balance)
    def get_balance(self):
        return self.__balance
    def deposit(self,amount):
        if amount>0:
            print("deposit succesfully")
            self.__balance+=amount
        else:
            return "Invalid ammount"
        return self.__balance
    
    def withdraw(self,amount):
         if amount<=0:
             return "Invalid amount"
         elif amount>self.__balance:
             return "Insufficient balance"
         else:
              self.__balance-=amount
         return f"withdraw succesfully :{amount}"
bank1=Bank_Account(24223,8000)
print("Account no=",bank1.acc_no)
print("Balance =",bank1.get_balance())
print("Balance=", bank1.deposit(2000))
print("Withdraw:", bank1.withdraw(2000))
print("Current Balance:", bank1.get_balance())
```

##### 7. Class Name: Car

##### Create class Car.

##### • __model

##### Provide
##### • get_model()

##### • set_year()
##### Validation:
##### Year cannot be greater than current year.

```python
from datetime import datetime
class Car:
    def __init__(self, model, year):
        self.__model = model
        self.__year = year
    def get_model(self):
        return self.__model
    def set_model(self, model):
        self.__model = model
        print("Model updated successfully.")
    def get_year(self):
        return self.__year
    def set_year(self, year):
        current_year = datetime.now().year
        if year <= current_year:
            self.__year = year
            print("Year updated successfully.")
        else:
            print("Invalid Year! Year cannot be greater than the current year.")

car1 = Car("Toyota Fortuner", 2022)

print("Model:", car1.get_model())
print("Year:", car1.get_year())
car1.set_model("Mahindra XUV700")
print("Updated Model:", car1.get_model())
car1.set_year(2025)
print("Updated Year:", car1.get_year())

car1.set_year(2030)
print("Final Year:", car1.get_year())
```

##### 10. Class Name: Account
##### Problem
##### Create class Account.
##### Explanation
##### Private attribute
##### __balance
##### Whenever balance changes,
##### store every balance value inside a class variable
##### Example
##### history = []
##### Methods
##### • deposit()
##### • withdraw()
##### Every successful update should append the latest balance to history.

```python
class Account:
    def __init__(self,name,balance):
        self.name=name
        self.__balance=balance
        self.history=[]
    def set_deposit(self,amount):
        self.history.append(self.__balance)
        if amount>0:
            print(f"Deposit Successfully {amount}")
            self.__balance+=amount
        else:
            return "Invalid balance"
    def get_balance(self):
        return self.__balance
    def set_withdraw(self,amount):
        if amount<0:
            print("Invalid amount")
        elif amount>self.__balance:
            print("Insufficient Balance")
        else:
            self.history.append(self.__balance)
            self.__balance-=amount
            self.history.append(self.__balance)
            print("Withdraw Successfully")
            print("Withdraw amount=",amount)
            print("Balance=",self.__balance)
user1=Account("Sameer",7000)
print(f"Name={user1.name}")
print(f"Balance={user1._Account__balance}")
print("*"*80)

user1.set_deposit(8000)
print("Balance=",user1.get_balance())
print("*"*80)

user1.set_withdraw(7500)
print("*"*80)

print("Balance history=",user1.history)

        
            
        
```

##### 16. Class Name: Shape
##### Problem
##### Create an abstract class Shape.
##### Explanation
##### Declare abstract methods
##### • calculate_area()
##### • calculate_perimeter()
##### Create subclasses like
##### • Circle
##### • Rectangle
##### Implement both methods.

```python
from abc import ABC, abstractmethod
class Shape(ABC):
    @abstractmethod
    def calculate_area(self):
        pass
    def calculate_perimeter(self):
        pass
class Circle:
    def __init__(self,radius):
        self.radius=radius
    def calculate_area(self):
        print("Area=",3.14* (self.radius**2))
    def calculate_perimeter(self):
        print("Perimeter=",2*3.14*self.radius)
class Rectraingle:
    def __init__(self,length,width):
        self.length=length
        self.width=width
    def calculate_area(self):
        print("Area=",self.length*self.width)
    def calculate_perimeter(self):
        print("Perimeter=",2*(self.length+self.width))  
print("Circle")
cr1=Circle(4)
cr1.calculate_area()
cr1.calculate_perimeter()
print("*"*90)
print("Rectraingle")
rec1=Rectraingle(24,14)
rec1.calculate_area()
rec1.calculate_perimeter()
```
