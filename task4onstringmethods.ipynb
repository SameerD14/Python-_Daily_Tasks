##### 16. (Hard) Take a sentence. Split into words, remove all words that start with vowel (a,e,i,o,u), then join remaining words.

```python
texts=input("Enter a sentence: ")
texts=texts.split()
word=[]
for text in texts:
    if not text.startswith(("a", "e", "i", "o", "u","A", "E", "I", "O", "U")):
        word.append(text)
print(" ".join(word))
```

##### 17. (Hard) Take a string. Replace all spaces with -, then split by -, and print only words with length > 4.


```python
text=input("Enter a string: ")
text=text.replace(" ","-")
text=text.split("-")
word=[]
for ch in text:
    if len(ch)>4:
        word.append(ch)
print(word)
        
```

##### 19. (Hard) Take a paragraph. Split it into sentences (using .), strip each sentence, and join them with "\n- ".

```python
text=input("Enter a text: ")
text=text.split(".")
word=[]
for ch in text:
    ch=ch.strip()
    if ch:
         word.append(ch)
print("\n-".join(word))
```

##### 20. (Hard) Take a string. Use replace(), strip(), split(), and join() together to clean and reformat it as a proper sentence.

```python
text="   hello---world---from---python   "
text=text.replace("---"," ")
text=text.strip()
text=text.split()
text=" ".join(text)
text=text.capitalize()
if not text.endswith("."):
    text += "."
print(text)

```
