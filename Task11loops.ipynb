##### 16. Use while loop to reverse a number (e.g., 1234 → 4321).

```python
num=int(input("Enter a number: "))
reverse=0
while num>0:
    digit=num%10
    reverse=reverse*10+digit
    num=num//10
print("reverse number=",reverse)
```

##### 19. Use while loop to check if a given number is prime or not.

```python
n=int(input("Enter a number:"))
count=0
i=1;
while i<=n:
    if n%i==0:
        count+=1
    i+=1
if count==2:
    print(f"{n} is a prime no")
else:
    print("Not a prime no")
```

# 5. (Hard) Create a number guessing game using while loop:
# • Generate a random number between 1 to 100.
# • Give the user 7 chances to guess.
# • After each guess, tell whether the number is "Too High" or "Too Low".
# • If user guesses correctly, print "Congratulations!" and stop.
# • If chances are over, reveal the number.

```python
import random
num=random.randint(1,100)
# print(num)
i=1
count=0
while i<=7:
    guess=int(input("Guess number between 1 to 100:"))
    count+=1
    if guess>num:
        print("Too High")
        if i==7:
            break
    elif guess<num:
        print("Too Low")
        if i==7:
            break
    elif  guess==num:
        print(f"guess is correct: {guess}")
        break;
    elif i==7:
        print("you lose the game..")
    i+=1
print("count=",count)
    
```
