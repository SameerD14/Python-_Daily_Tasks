```python
# 5. Arithmetic Operators
# Python
# Take two numbers from user and print:
# Sum, Difference, Product, Division, Floor Division, Modulus, Power
num1=int(input("Enter a number 1 : "))
num2=int(input("Enter a number 2 : "))
Sum=num1+num2
Difference=num1-num2
Product=num1*num2
Division=num1/num2
Floor_Division=num1//num2
Modulus=num1%num2
Power=num1*num2
print("Sum=",Sum)
print("Difference=",Difference)
print("Product=",Product)
print("Division=",Division)
print("Floor_Division=",Floor_Division)
print("Modulus=",Modulus)
print("Power=",Power)
```

```python
# 6. Comparison Operators
# Python
# Take two numbers and check:
# Equal, Not Equal, Greater, Smaller, Greater or Equal, Smaller or Equal
num1=int(input("Enter a number 1 : "))
num2=int(input("Enter a number 2 : "))
print("Equal=",num1==num2)
print("Not Equal=",num1!=num2)
print("Greater=",num1>num2)
print("Smaller=",num1<num2)
print("Greater or Equal=",num1>=num2)
print("Smaller or Equal=",num1<=num2)
```

```python
# 9. Simple Calculator
# Python
# Create a simple calculator that takes two numbers and one operator (+,-,*,/,//,%,**)
# from the user and shows the result.
num1=int(input("Enter a number 1 : "))
num2=int(input("Enter a number 2 : "))
Operator=input("Enter a operator: ")
if Operator=="+":
   print("Sum: ",num1+num2)
if Operator=="-":
   print("Sub: ",num1-num2)

if Operator=="*":
   print("Mul: ",num1*num2)
if Operator=="/":
   print("Div: ",num1/num2)

if Operator=="%":
   print("Modulo: ",num1%num2)

if Operator=="//":
   print("floor_division: ",num1//num2)

if Operator=="**":
   print("power: ",num1**num2)

```

```python
# 11. Area and Perimeter Calculator
# Python
# Take radius of circle from user and calculate:
# Area and Circumference of circle
# (Use pi = 3.14)
r=float(input("Enter a radius: "))
Area=3.14*r**2
Circumference=2*3.14 *r
print("Area of Circle= ",Area)
print("Circumference of Circle= ",Circumference)
```

```python
# 12. Student Grade Calculator
# Python
# Take marks of 5 subjects.
# Calculate percentage and print grade:
# >=90 → A+, >=80 → A, >=70 → B, >=60 → C, else → Fail
sub1=int(input("Enter a sub1 : "))
sub2=int(input("Enter a sub2 : "))
sub3=int(input("Enter a sub3 : "))
sub4=int(input("Enter a sub4 : "))
sub5=int(input("Enter a sub5 : "))
total=sub1+sub2+sub3+sub4+sub5
per=total*100/500
if per>=90:
    print("Grade A+ ,percentage is ",per)
elif per>=80:
    print("Grade A ,percentage is ",per)
elif per>=70:
    print("Grade B ,percentage is ",per)
elif per>=60:
    print("Grade C,percentage is ",per)
else:
    print("Fail")
```

```python
# 13. Swap Two Variables
# Python
# Take two numbers from user and swap them using:
# a) Third variable
# b) Without third variable (Python way)
num1=int(input("Enter a number1: "))
num2=int(input("Enter a number2: "))
temp=num1
num1=num2
num2=temp
print("Swap num1=num2: ",num1)
print("Swap num2=num1: ",num2)
```

```python
# b) Without third variable (Python way)
num1=int(input("Enter a number1: "))
num2=int(input("Enter a number2: "))
num1,num2=num2,num1

print("Swap num1=num2: ",num1)
print("Swap num2=num1: ",num2)
```
