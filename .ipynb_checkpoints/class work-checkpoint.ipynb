```python
# copy
sets={4,6,7,3,6,1}
setscopy=sets.copy()
setscopy
```

```python
# Union (|)
sets={4,6,7,3,6,1}
set1={3,3,5,2,5,7,87}

# a=sets | set1
a=sets.union( set1)
a
```

```python
# INTERSECTION (&)
sets={4,6,7,3,6,1}
set1={3,3,5,2,5,7,87}
# a=sets & set1
a=sets.intersection(set1)
a
```

```python
# Difference (-)
sets={4,6,7,3,6,1}
set1={3,3,5,2,5,7,87}
# a=sets - set1
a=sets.difference(set1)
a
```

```python
# Symetric Difference (^)
sets={4,6,7,3,6,1}
set1={3,3,5,2,5,7,87}
a=sets ^ set1
a
```
