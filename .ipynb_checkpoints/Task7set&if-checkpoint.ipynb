##### 5. (Medium) Create two sets A = {1, 2, 3, 4} and B = {3, 4, 5, 6}. Print union, intersection, and difference.

```python
A = {1, 2, 3, 4} 
B = {3, 4, 5, 6}
print(A,"\n",B)
print("union=",A.union(B))
print("Intersection=",A.intersection(B))
print("Difference=",A.difference(B))
print("Symmetric Difference=",(A ^ B))
```

##### 3. (Hard) Take three sides of a triangle as input. Check if they can form a valid triangle (sum of any two sides must be greater than the third side). If valid, also print the type (Equilateral, Isosceles, or Scalene).

```python
a=int(input("Enter a side a =" ))
b=int(input("Enter a side b =" ))
c=int(input("Enter a side c =" ))
if (a+b)>=c and (b+c)>=a and (a+c)>=b:
    print("Valid Traingle")
    if (a==b and a==c) or (b==c and a==c) or (a==b and b==c):
        print("Equilateral Traingle..")
    elif a==b or b==c or a==c:
        print("Isoceles Traingle")
    else:
        print("Scalene Traingle")
else:
        print("Invalid Traingle")

    
```

##### 4. (Hard) Take a student's marks in 5 subjects. Calculate percentage and print grade according to these rules:
##### • 90 or above → A+
##### • 80–89 → A
##### • 70–79 → B
##### • 60–69 → C
##### • Below 60 → Fail Also print "Passed with Distinction" if percentage >= 75.

```python
a=int(input("Enter a marks sub1: "))
b=int(input("Enter a marks sub2: "))
c=int(input("Enter a marks sub3: "))
d=int(input("Enter a marks sub4: "))
e=int(input("Enter a marks sub5: "))
total=a+b+c+d+e
per=total*100/500
if per>=90:
    print("A+")
elif per>=80:
    print("A")
elif per>=70:
    print("B")
elif per>=60:
    print("C")
else:
    print("Fail")
if per>=75:
    print("Passed With Distinction")
```
