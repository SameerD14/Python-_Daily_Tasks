# Case Conversion Questions
# 6. Take a string from user and print it in:
### • Uppercase
### • Lowercase
### • Title Case
### • Capitalized (only first letter capital)


```python
text=input("Enter a message: ")
print("Upper Case =" ,text.upper())
print("Lower Case =", text.lower())
print("Title case =",text.title())
print("Captalized =",text.capitalize())
```

# 7. Take a sentence and convert it to title case. Then print the original and converted version.

```python
text=input("Write a Message: ")
converted_text=text.title()
print("Original text=",text)
print("Converted text=",converted_text)
```

# 8. Take a mixed case string and check if lower() and upper() methods change the string permanently or not.


```python
text=input("Take a input Mixed case String: ")
print("check lower=",text.islower())
print("check Upper=",text.isupper())
print(text)
```

##### 13. Take a password as input and check:
###### • It should be alphanumeric
###### • It should be minimum 8 characters long Print "Valid Password" or "Invalid Password".

```python
text=input("Enter  a message=")
result="Valid" * (text.isalnum() and len(text)>=8) or "Invalid Password"
print(result)
```

###### 15. Take a string and a word. Use find() to check if the word exists. If not found, print -1.

```python
text=input("Enter a string= ")
word=input("Enter a word= ")
result="exist" * (text.find(word)!=-1 ) or "not exist"
print(result)
```

###### . Take a sentence and find the position of the word "Python" using index().

###### 17. Take a string. Use both find() and index() to search for a character that does not exist. Observe the difference.

```python
text=input("Enter a String: ")
word=input("Enter a word= ")
print("Position of word=",text.find(word))
print("position of word= ",text.index(word))
```

###### Mixed Questions 18. Take your full name.• Convert it to uppercase• Find the position of first space• Print the first name and last name separately.

```python
text=input("Enter a full name: ")
print("Converting into Uppercase :",text.upper())
print("Find the position of first space :",text.find(" "))
first_name=text.find(" ")
last_name=text.rfind(" ")
print(text[:first_name],text[last_name:])
```

###### 20. Take any string from user. • Print the string in reverse order using slicing • Convert it to title case • Check if the reversed string is same as original (Palindrome check)

```python
text=input("Enter a string: ")
reverse=text[::-1]
print("Original string: ",text)
print("string in reverse order:",reverse)
print("conver it into Title case: ",reverse.title())
result="palindrome" * (text==reverse) or "NOT Palindrome"
print("Check the string is palidome or not: "result)

```
