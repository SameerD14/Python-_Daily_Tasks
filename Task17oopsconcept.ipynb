##### 1. Create a list original = [1, 2, [3, 4], 5]. Make a shallow copy named shallow and change shallow[2][0] = 99. Print both lists and explain the result.

```python
import copy
original = [1, 2, [3, 4], 5]
print("Original list=",original)

shallow_copy=copy.copy(original)
shallow_copy[2][0] =99
print("Original list=",original)
print("shallow list=",shallow_copy)


```

##### 2. Do the same as above but use deepcopy instead of shallow copy. Show the difference.

```python
original = [1, 2, [3, 4], 5]
print("Original list=",original)
deep_copy=copy.deepcopy(original)
deep_copy[2][0] =99
print("Original list=",original)
print("deep list=",deep_copy)
```

##### 3. Write a program to demonstrate shallow copy using copy.copy() on a nested list.

```python
import copy
original = [[1, 2], [3, 4], [5, 6]]
shallow=copy.copy(original)
print("Befor Modification")
print("Original copy=",original)
print("Shallow copy=",shallow)
shallow[0][0]=33
print("After modification")
print("Original copy=",original)
print("Shallow copy=",shallow)
```

##### 4. Write a program to demonstrate deep copy using copy.deepcopy() on the same nested list

```python
import copy
original = [[1, 2], [3, 4], [5, 6]]
deep=copy.deepcopy(original)
print("Befor Modification")
print("Original copy=",original)
print("Deep copy=",deep)
deep[0][0]=33
print("After modification")
print("Original copy=",original)
print("Deep copy=",deep)
```

##### 5. Create a nested dictionary and perform shallow copy. Change a value in the inner dictionary and show the effect on original.

```python
import copy
# Original nested dictionary
dict1 = {"student1":{ "name": "Sameer","age": 22},
    "student2":{"name": "Sumit","age": 27},
    "student3": {"name": "Sujal","age": 21}}
# Shallow copy
shallow = copy.copy(dict1)
print("Before Modification")
print("Original =", dict1)
print()
print("Shallow  =", shallow)
# Change inner dictionary
shallow["student1"]["age"] = 25
print("\nAfter Modification")
print("Original =", dict1)
print()
print("Shallow  =", shallow)
```

##### 6. Car Speed Attributes • brand • speed
##### Method increase_speed(value) Maximum speed should never exceed 200 km/h.

```python
class Car:
    def __init__(self,brand,speed):
        self.brand=brand
        self.speed=speed
    def increase_speed(self,increase):
        if self.speed+increase<=200:
            self.speed+=increase
            print("Current Speed =",self.speed,"km/hr")
        else:
            self.speed = 200
            print("Maximum speed is 200 km/h.")
            print("Current Speed =", self.speed, "km/h")
car1=Car("Lambo",150)
car1.increase_speed(30)
car1.increase_speed(50)
```

##### 8. Movie Ticket Attributes • movie_name • seats
##### Method book_ticket(number) Booking should happen only if seats are available.

```python
class Movie:
    def __init__(self,movie_name,seats):
        self.movie_name=movie_name
        self.seats=seats
    def book_ticket(self,number):
        if number<=self.seats:
            print(f"Your {number} ticket is book")
            self.seats-=number
            print("Remaining seats=",self.seats)
        else:
            print("Remaining seats=",self.seats)
            print(f"{number} Seat is not available")
s1=Movie("Pushpa",100)
s1.book_ticket(3)
s1.book_ticket(29)
s1.book_ticket(70)

            
    
```

##### 9. Library
##### Attributes • book_name • available_copies
##### Method borrow_book() Reduce copy only if available.

```python
class Library:
    def __init__(self,book_name,available_copies):
        self.book_name=book_name
        self.available_copies=available_copies
    def borrow_book(self,copy):
        if copy==self.book_name:
            if self.available_copies>0:
                self.available_copies-=1
                print(f"Borrowed {self.book_name} Remaining books is {self.available_copies}")
            else:
                print("No copies available to borrow")
        else:
            print("Book is not found")
book1=Library("Alexander",20)
book1.borrow_book("Alexander")
book2=Library("Alexander",20)
book2.borrow_book("data science")
book1=Library("Alexander",20)
book1.borrow_book("Alexander")


        
```

##### 10. Product Attributes • product_name • quantity
##### Method sell(quantity) Quantity should never become negative.

```python
class Product:
    def __init__(self,product_name,quantity):
        self.product_name=product_name
        self.quantity=quantity
    def sell(self,quantity):
        if self.quantity>quantity:
            self.quantity-=quantity
            print(f"You buy {quantity} {self.product_name} Sucessfully")
            print("Now Available Quantity=",self.quantity)
        else:
            print("product is not available")
product1=Product("pen",100)
product1.sell(70)
```

##### 13. Fuel Tank Attributes • capacity • current_fuel Method fill(amount) Fuel should never exceed tank capacity.

```python
class FuelTank:
    def __init__(self,capacity,current_fuel):
        self.capacity=capacity
        self.current_fuel=current_fuel
    def fill(self,amount):
        if self.capacity>=self.current_fuel+amount:
            self.current_fuel+=amount
            print(f"{amount} liters added succesfully")
            print(f"Current fuel={self.current_fuel} liters")
        else:
            print("Tank capacity exceeds")
            print("Maximum capacity is :",self.capacity)
            print("Current Fuel capacity  is :",self.capacity-self.current_fuel)
f1=FuelTank(4000,2000)
f1.fill(2000)
f1.fill(200)
```

##### 14. Phone Battery Battery starts at some percentage. Method charge(value) Battery should never exceed 100%.

```python
class Phone:
    def __init__(self,current_battery):
        self.current_battery=current_battery
    def charge(self,value):
        if 0<=self.current_battery+value<=100:
            self.current_battery+=value
            print("Now battery percentage is ",self.current_battery)
        else:
            self.current_battery = 100
            print("Battery is fully charged!")
            print("Battery Percentage:", self.current_battery)
phone1=Phone(40)
phone1.charge(30)
phone1.charge(10)
phone1.charge(30)
```

##### 15. Wallet Attributes • owner • balance Method purchase(amount) Purchase allowed only if enough money exists.

```python
class Wallet:
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance
    def purchase(self,amount):
        if self.balance>=amount:
            print("Purchase Succeed")
            print("Balance is ",self.balance-amount)
        else:
            print("You have insufficient money")
            print("Balance is ",self.balance)
wallet1=Wallet("Sameer",8000)
wallet1.purchase(5700)
```
